```python
class Solution:

    def inversionCount(arr, n):
        
        return mergeSort(arr, n-1)
    
        
    def mergeSort(self, arr, l, r):
        inv = 0
        if l < r:
            mid = (l + r) // 2
            
            inv += self.mergeSort(arr, l, mid)
            inv += self.mergeSort(arr, mid+1, r)
            
            inv += self.merge(arr, l, mid, r)
            
        return inv
        
        
    def merge(self, arr, l, mid, r):
        inv = 0
        
        n1 = mid+1 - l
        n2 = r - mid
        
        a = [0] * (n1)
        b = [0] * (n2)
        
        for i in range(n1):
            a[i] = arr[l + i]
            
        for j in range(n2):
            b[j] = arr[mid+1 + j]
            
        
        i = 0
        j = 0
        k = l
        while(i < n1 and j < n2):
            if a[i] <= b[j]:
                arr[k] = a[i]
                k += 1
                i += 1
                
            else:
                arr[k] = b[j]
                k += 1
                j += 1
                inv += n1 - i
                
        while(i < n1):
            arr[k] = a[i]
            k += 1
            i += 1
            
        while(j < n2):
            arr[k] = b[j]
            k += 1
            j += 1
        
        return inv
    
    
    
arr = [2, 5, 8, 1, 5, 3, 9]

n = len(arr)

ob = Solution()

count = Solution.inversionCount(arr, n)

print(count)
```
